```python
import statsmodels.formula.api as smf
```

```python
model = smf.ols("Salary~YearsExperience",data=Salary).fit()
Salary
```

```python
import seaborn as sns
sns.regplot(x='YearsExperience',y='Salary',data=Salary)
```

```python
model.params
```

```python
Salary.head()
```

```python
model.fittedvalues
```

```python
Salary['Predicted']=model.fittedvalues
Salary
```

```python
model.resid
```

```python
Salary['Error']=model.resid
Salary
```

```python
Salary['Errors_Sq']=(Salary['Error'])**2
Salary
```

```python
#Intercept          25792.200199
#YearsExperience     9449.962321
Salary['Salary_Hike']= 25792.200199 + (9449.962321 * Salary['YearsExperience']+Salary['Error'])
Salary[0:5]
```

```python
Mean_Squared_Error=Salary['Errors_Sq'].mean()
Mean_Squared_Error
```

```python
model.rsquared
```

```python
model.pvalues
```

```python
Salary['SQRT_YearsExperience']=np.sqrt(Salary['YearsExperience'])
Salary
```

```python
import statsmodels.formula.api as smf
```

```python
model = smf.ols("Salary~SQRT_YearsExperience",data=Salary).fit()
Salary
```

```python
model.params
```

```python
model.fittedvalues
```

```python
model.resid
```

```python
Salary['SQRT_Predicted']=model.fittedvalues
Salary['SQRT_Errors']=model.resid
Salary.head()
```

```python
Salary['SQRT_Errors_Sq']=Salary['SQRT_Errors']**2
Salary.head()
```

```python
Salary['Salary_Hike']= 25792.200199 + (9449.962321 * Salary['YearsExperience']+Salary['Error'])
Salary[0:5]
```

```python
#Intercept              -16055.769117
#SQRT_YearsExperience    41500.680583
Salary['SQRT_Y']=-16055.769117 + (41500.680583 * Salary['SQRT_YearsExperience']+Salary['SQRT_Errors'])
Salary[0:5]
```

```python
MSE=Salary['SQRT_Errors_Sq'].mean()
MSE
```

```python
import statsmodels.formula.api as smf
```

```python
model=smf.ols('SortingTime~DeliveryTime',data=Time).fit()
```

```python
sns.regplot(x='DeliveryTime',y='SortingTime',data=Time)
```

```python
model.params
```

```python
model.fittedvalues
```

```python
Time
```

```python
Time['Predicted']=model.fittedvalues
Time
```

```python
model.resid
```

```python
Time['Errors']=model.resid
Time
```

```python
Time['Errors_Sq']=Time['Errors']**2
Time
```

```python
#Intercept      -0.756673
#DeliveryTime    0.413744
Time['Y']=-0.756673+(0.413744*Time['DeliveryTime']+Time['Errors'])
Time
```

```python
Mean_Squared_Error=Time['Errors_Sq'].mean()
Mean_Squared_Error
```
